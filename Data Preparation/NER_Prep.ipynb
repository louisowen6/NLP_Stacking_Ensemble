# NER Tagging

```python
# set up the client
print('---')
print('starting up Java Stanford CoreNLP Server...')
# set up the client
client=CoreNLPClient(annotators=['tokenize','ner'], timeout=30000, memory='16G')
```

```python
def ner(text,tag):
    exist=0
    ann = client.annotate(text)
    iter_sent=range(len(ann.sentence))
    for i in iter_sent:
        sentence=ann.sentence[i]
        for j in range(len(sentence.token)):
            if sentence.token[j].ner==tag.upper():
                exist=1
    return exist
```

```python
ner_tag_list=['DATE', 'DURATION', 'LOCATION', 'MONEY', 'NUMBER', 'ORDINAL', 'ORGANIZATION', 'PERCENT', 'PERSON', 'SET', 'TIME']
for tag in ner_tag_list:
    df['NER_'+tag]=df['clean_text'].apply(lambda x: ner(x,tag) if type(x)==str else np.nan)
```
